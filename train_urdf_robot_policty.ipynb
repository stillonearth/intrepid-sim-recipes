```python
import torch as th
```

```python
from stable_baselines3 import SAC, PPO
from stable_baselines3.common.monitor import Monitor
```

```python
wrapped_env = Monitor(env)
```

```python
policy_kwargs = dict(
    activation_fn=th.nn.ReLU,
    net_arch=dict(pi=[512, 512, 512, 512], qf=[512, 512, 512, 512])
)
```

```python
model = SAC("MlpPolicy", wrapped_env, verbose=1, policy_kwargs=policy_kwargs,  tensorboard_log="./flamingo/")
```

```python
model.learn(total_timesteps=5000, log_interval=10, progress_bar=False)
```

```python
model.save('logs/lamingo-ppo')
```
